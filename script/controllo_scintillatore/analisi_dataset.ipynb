Grafico dell'andamento della tensione in output del power supply con collegato il setup completo

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
dataset=pd.read_csv('data_IV_scintillatore.csv')
print(dataset)
```

```python
ordered_dataset = dataset.sort_values(by=dataset.columns[0])
plt.clf()
plt.plot(ordered_dataset.iloc[:,0],ordered_dataset.iloc[:,1], color='#000000')
plt.xlabel("Voltage (V)")
plt.ylabel("Current (A)")
plt.show()
```
